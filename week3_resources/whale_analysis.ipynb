```python
## From 1. above; I want to calculate and plot the rolling standard deviation for the S&P 500 using a 21-day window
sp500_pct.rolling(window=21).std().plot(figsize = (20,5))
```

```python
## I'm also calculating the rolling standard deviation for all the portfolios using the same 21-day window.
## This is just as a sanity-check for when I do correlations in the next step.

joined_rolling21std = joined_df.rolling(window=21).std()

# Plotting the rolling standard deviation.
joined_rolling21std.plot(figsize=(20,10))
```

```python
## From 2. above, I want to caluclate the correlation of each stock to see which portfolios *may* track the S&P Benchmark_Close.
# Calculating the correlation.
joined_corr = joined_df.corr()
joined_corr
```

```python
# Now we want to display the correlations as a heat map.
sns.heatmap(joined_corr, vmin=-1, vmax=1);
```

```python
## From 3. above, I need to calculate the rolling beta between one portfolio and the S&P 500.
## From the beta student do, it looks like I need to figure out the variance and covariance of a portfolio and the S&P.
## I'll use Algo 1, and call my variance and covariance variables "algovar" and "algocov".

algovar = joined_df['Algo 1'].var()
algovar
```

```python
algocov = joined_df['Algo 1'].cov(joined_df['Benchmark_Close'])
algocov
```

```python
## From the student do, I want to caluclate the beta of Algo 1, and I'm calling this variable "algobet".

algobet = algocov / algovar
algobet
```

```python
## Now we need to plot the beta.
## In the student do, we showcased beta vs. correlation by plotting a scatterplot using the seaborn library and fitting a regression line.
sns.lmplot(x='Benchmark_Close', y='Algo 1', data=joined_df, aspect=1.5, fit_reg=True);
```

### Rolling Statistics Challenge: Exponentially Weighted Average

### Sharpe Ratios
Investment managers and their institutional investors look at the return-to-risk ratio, not just the returns.<br>
After all, if you have two portfolios that each offer a 10% return, yet one is lower risk, you would invest in the lower-risk portfolio, right?<br>

```python
# Using the daily returns, calculate and visualize the Sharpe ratios using a bar plot.
```

```python
## I'm working on this across multiple computers over time, so I moved the risk_free_rate.csv file into my working dir.
## Using the Pathlib libary to set the path to the Risk Free Rate CSV (I'm naming the variable "rfr_path").
rfr_path = Path('risk_free_rate.csv')

## I'm going to read the csv file into the notebook using the same index and inference as the other dfs.
rfr_df = pd.read_csv(rfr_path, index_col = "Date", infer_datetime_format = True, parse_dates = True)
rfr_df
```

```python
# Joining my returns DataFrame to the original returns DataFrame which includes the Risk Free Rate.

joined_withrfr = pd.concat([rfr_df, joined_df], axis="columns", join="inner")
joined_withrfr
```

```python
# Calculating Annualzied Sharpe Ratios (lots of help from Google).
## I effectively copy/pasted this from Google and had to swap the dataframe names and column names a bit until it worked.
## The formula sets the risk free rate to "0 risk" (it's in the name), and compares the portfolios to that rate, expressed as a %.

joined_sharpe = ((joined_withrfr.mean()-joined_withrfr['rf_rate'].mean()) * 252) / (joined_withrfr.std() * np.sqrt(252))

# Sorting the portfolios by Sharpe Ratio to compare them to Risk Free Rate (rf_rate).

joined_sharpe = joined_sharpe.sort_values(ascending=False)
joined_sharpe
```

```python
# Visualize the sharpe ratios as a bar plot.
joined_sharpe.plot(figsize=(20,5), kind="bar")
```

# Random Forests

```python
# Initial imports
import pandas as pd
from pathlib import Path
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
%matplotlib inline

import numpy as np
from sklearn                        import metrics, svm
from sklearn.linear_model           import LinearRegression
from sklearn.linear_model           import LogisticRegression
from sklearn.tree                   import DecisionTreeClassifier
from sklearn.neighbors              import KNeighborsClassifier
from sklearn.discriminant_analysis  import LinearDiscriminantAnalysis
from sklearn.naive_bayes            import GaussianNB
from sklearn.svm                    import SVC

# Needed for decision tree visualization
import pydotplus
from IPython.display import Image
```

## Loading and Preprocessing Loans Encoded Data

```python
# Loading data
file_path = Path("dfs_new.csv")
dfs_new = pd.read_csv(file_path, parse_dates=True, infer_datetime_format=True, index_col="Unnamed: 0")
dfs_new
```

```python
## What are the columns?
dfs_new.columns
```

```python
# Making the TSLA DF
tsla_df= pd.DataFrame(dfs_new, columns = ['Tesla,_Inc.views','TSLA_RSI', 'TSLA','TSLA_bollinger_band_mid_band','TSLA_bollinger_std','TSLA_bollinger_upper_band','TSLA_bollinger_lower_band'])
```

```python
# Renaming columns.
tsla_df.index.rename("Date", inplace=True)
```

```python
tsla_df
```

```python
## Filling the NaN values (first 15 days) with 0.

tsla_df = tsla_df.fillna(0)
tsla_df
```

```python
## I need to get a pct change from the TSLA closing prices so I can get a + or - on the day.
## We need a margin of error ("x%" = 0 to compensate for flat days or $0.01 or$0.02, for example); isolates large swings.
## I need to convert the + or - on the day to a binary input (trade on +1 or -1, no action on 0).
## I need to make the binary data the y in my features set.
```

```python
## Getting % daily change from daily close column (named as ticker).

pct_chg = tsla_df['TSLA'].pct_change()
pct_chg
```

```python
## Next, I need to get the data in pct_change to populat the "Daily_Change" column in my DF.
```

```python
tsla_df["Daily Change"] = pct_chg
tsla_df
```

```python
## Now, I want to make a signal using the % change; higher than 0.02
print(tsla_df['Daily Change'])
```

```python
## I am going to construct a trading signal based on +/- 2% daily change.
## Construct a trading signal using the Daily Change value.
## We want a signal when the stock is going up, going down, or trading sideways. up/down/hold.

tsla_df['up_signal'] = np.where(tsla_df['Daily Change'] > 0.02, 1.0, 0.0)
tsla_df['down_signal'] = np.where(tsla_df['Daily Change'] < -0.02, -1.0, 0.0)
```

```python
## I want to COMBINE the up and down signals into 1 binary value.
tsla_return = tsla_df['up_signal'] + tsla_df['down_signal']
tsla_return
```

```python
## From 15.1 (rinse and repeat)

# Set short and long windows
short_window = 10
long_window = 20

# Construct a `Fast` and `Slow` Exponential Moving Average from short and long windows, respectively
tsla_df['fast_close'] = tsla_df['TSLA'].ewm(halflife=short_window).mean()
tsla_df['slow_close'] = tsla_df['TSLA'].ewm(halflife=long_window).mean()

# Construct a crossover trading signal
tsla_df['crossover_long'] = np.where(tsla_df['fast_close'] > tsla_df['slow_close'], 1.0, 0.0)
tsla_df['crossover_short'] = np.where(tsla_df['fast_close'] < tsla_df['slow_close'], -1.0, 0.0)
tsla_df['crossover_signal'] = tsla_df['crossover_long'] + tsla_df['crossover_short']

tsla_df.head()
```

```python
# Plotting the exponential moving averages of each ticker's closing prices

tsla_df[['TSLA', 'fast_close', 'slow_close']].plot(figsize=(20,10))
```

```python
tsla_df = tsla_df.fillna(0)
```

```python
## We want to bin the TSLA_RSI then standardize scale it.
## use pd.qcut; https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.qcut.html

tsla_df['RSI_binned']=pd.qcut(tsla_df['TSLA_RSI'],10,labels=False)
```

```python
## We want to bin the Tesla,_Inc.views and then standardize scale it.

tsla_df['views_binned']=pd.qcut(tsla_df['Tesla,_Inc.views'],10,labels=False)
```

```python
# Define features set
X = tsla_df[['crossover_signal','RSI_binned','views_binned']].copy()
X.head()
```

```python
## Looking better with dates. No NaN, values pick up after 15th entry.
```

```python
## Looks good-- moving on.
```

```python
# Define target vector
## Looks like we're after the TSLA close price.

y = tsla_return
y
```

```python
# Splitting into Train and Test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
# Creating StandardScaler instance
scaler = StandardScaler()
```

```python
# Fitting Standard Scaller
X_scaler = scaler.fit(X_train)
```

```python
# Scaling data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Fitting the Random Forest Model

```python
# Fitting the model

rf_model = RandomForestClassifier(n_estimators=100, random_state=1)
rf_model.fit(X_train, y_train)
```

```python
# Make a prediction of "y" values from the X_test dataset
predictions = rf_model.predict(X_test)

# Assemble actual y data (Y_test) with predicted y data (from just above) into two columns in a dataframe:
Results = y_test.to_frame()
Results["Predicted Value"] = predictions
Results
```

## Making Predictions Using the Random Forest Model

```python
# Making predictions using the testing data
predictions = rf_model.predict(X_test_scaled)
```

## Model Evaluation

```python
# Calculating the confusion matrix
cm = confusion_matrix(y_test, predictions)
cm_df = pd.DataFrame(
    cm, index=["Actual -1", "Actual 0", "Actual 1"], columns=["Predicted -1", "Predicted 0", "Predicted 1"]
)

# Calculating the accuracy score
acc_score = accuracy_score(y_test, predictions)
```

```python
# Displaying results
print("Confusion Matrix")
display(cm_df)
print(f"Accuracy Score : {acc_score}")
print("Classification Report")
print(classification_report(y_test, predictions))
```

## Feature Importance

```python
# Random Forests in sklearn will automatically calculate feature importance
importances = rf_model.feature_importances_
```

```python
# We can sort the features by their importance
sorted(zip(rf_model.feature_importances_, X.columns), reverse=True)
```
